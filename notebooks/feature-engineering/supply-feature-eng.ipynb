```python
#  Import da base supply

df_supply_o = pd.read_csv('/content/drive/Shareddrives/grupo4-rappi-hour/bases-rappi/supply-def.csv')
```

```python
#  Drop de colunas não importantes para a feature
#  Já estão em outras bases

df_supply = df_supply_o.drop(columns=['CITY', 'DATE', 'WEEK', 'CREATED_CARD', 'LEVEL_NAME_2', 'HAVE_CARD', 'TRANSPORT_MEDIA_TYPE'])
```

```python
#  Soma dos pedidos realizados e das horas ativas no aplicativo de cada entregador
#  Tabela organizada com uma linha para cada entregador registrado

df_supply = df_supply.groupby(['STOREKEEPER_ID']).sum().reset_index()
```

```python
#  CRIAÇÃO DA FEATURE
#
#  Criação da nova coluna "ORDERS_PER_SUPPLY_HOURS" - média de pedidos por hora 
# de atividade no aplicativo (por entregador)
#  Representa a produtividade média dos entregadores

df_supply['ORDERS_PER_SUPPLY_HOURS'] = df_supply['NUM_ORDERS'] / df_supply['SUPPLY_HOURS']
```

```python
#  Remoção das colunas do número de pedidos e do número de horas ativas
#  Não é tão relevante para o modelo

df_supply = df_supply.drop(columns=['NUM_ORDERS', 'SUPPLY_HOURS'])
```

```python
#  Compatibilidade da chave de merge (ID) com da base de supply com a base consolidada

df_supply = df_supply.rename(columns={'STOREKEEPER_ID': 'ID'})
```
