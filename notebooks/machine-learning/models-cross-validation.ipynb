```python
df.info()
```

# cross validation target / features


```python
x = df.drop(columns="IS_ACTIVE") # features
y = df["IS_ACTIVE"] # target
```

```python
y.squeeze()
```

```python
scoring = ['precision_macro', 'recall_macro']
```

```python
from sklearn.model_selection import cross_validate
```

# **Models**

## LogReg cross validation

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
```

```python
from sklearn.model_selection import cross_validate

logreg_scores = cross_validate(logreg, x, y, cv=5, scoring=scoring, error_score='raise')
```

```python
logreg_scores
```

## adaboost cross validation

```python
from sklearn.ensemble import AdaBoostClassifier

adaboost = AdaBoostClassifier()
```

```python
ada_scores = cross_validate(adaboost, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
ada_scores
```

## ranfor cross validation

```python
from sklearn.ensemble import RandomForestClassifier

ranfor = RandomForestClassifier()
```

```python
ranfor_scores = cross_validate(ranfor, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
ranfor_scores
```

## sgd cross validation

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier()
```

```python
sgd_scores = cross_validate(sgd, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
sgd_scores
```

## svm cross validation

```python
from sklearn import svm

svm = svm.SVC()
```

```python
svm_scores = cross_validate(svm, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
svm_scores
```

## knn cross validation

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
```

```python
scores_knn = cross_validate(knn, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
scores_knn
```

## lgbm cross validation


```python
import lightgbm as lgb

lgb = lgb.LGBMClassifier()
```

```python
scores_lgb = cross_validate(lgb, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
scores_lgb
```

# **Cluster Resample**

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=13)
x_resampled, y_resampled = rus.fit_resample(x, y)
```

## LogReg resampled cross validation

```python
from sklearn.model_selection import cross_validate

scores_logreg_resampled = cross_validate(logreg, x_resampled, y_resampled, cv=5, scoring=scoring, error_score='raise')
```

```python
scores_logreg_resampled
```

```python
scores_lgb_resampled = cross_validate(lgb, x_resampled, y_resampled, cv=5, error_score='raise', scoring=scoring)
```
