## cross validation target / features


```python
x = df.drop(columns="IS_ACTIVE") # features
y = df["IS_ACTIVE"] # target
```

```python
y.squeeze()
```

```python
from sklearn.model_selection import cross_validate

scoring = ['precision_macro', 'recall_macro']
```

## ranfor cross validation

```python
from sklearn.ensemble import RandomForestClassifier

ranfor = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=None, max_features='auto',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=100,
                       n_jobs=-1, oob_score=False, random_state=5330, verbose=0,
                       warm_start=False)
```

```python
ranfor_scores = cross_validate(ranfor, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
ranfor_scores
```

## lgbm cross validation


```python
import lightgbm as lgb

lgb = lgb.LGBMClassifier(boosting_type='gbdt', class_weight=None, colsample_bytree=1.0,
               importance_type='split', learning_rate=0.1, max_depth=-1,
               min_child_samples=20, min_child_weight=0.001, min_split_gain=0.0,
               n_estimators=100, n_jobs=-1, num_leaves=31, objective='multiclass',
               random_state=5330, reg_alpha=0.0, reg_lambda=0.0, silent='warn',
               subsample=1.0, subsample_for_bin=200000, subsample_freq=0)
```

```python
scores_lgb = cross_validate(lgb, x, y, cv=5, error_score='raise', scoring=scoring)
```

```python
scores_lgb
```

## cluster resample

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=13)
x_resampled, y_resampled = rus.fit_resample(x, y)
```

## ranfor resampled cross validation

```python
scores_ranfor_resampled = cross_validate(ranfor, x_resampled, y_resampled, cv=5, error_score='raise', scoring=scoring)
```

```python
scores_ranfor_resampled
```

## lgbm resampled cross validation


```python
scores_lgb_resampled = cross_validate(lgb, x_resampled, y_resampled, cv=5, error_score='raise', scoring=scoring)
```

```python
scores_lgb_resampled
```
